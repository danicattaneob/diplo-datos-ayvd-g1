```
import seaborn as sns # for data visualization
import matplotlib.pyplot as plt # for data visualization
import pandas as pd
import io
import requests
```

```
url = "https://raw.githubusercontent.com/DiploDatos/AnalisisYCuracion/master/input/pasos.json"
s = requests.get(url).content
dataset_pasos = pd.read_json(io.StringIO(s.decode('utf-8')))
```

```
dataset_pasos.head()
```

```
dataset_grouped = dataset_pasos.groupby('province').count()
dataset_grouped=dataset_grouped.reset_index()
dataset_grouped
```

```
plt.figure(figsize=(16,16))
sns.barplot(y='province', x='location', data=dataset_grouped, orient='h')
```
