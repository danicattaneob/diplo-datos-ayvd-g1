```
import seaborn as sns # for data visualization
import matplotlib.pyplot as plt # for data visualization
import pandas as pd
import io
import requests
```

```
url = "https://raw.githubusercontent.com/DiploDatos/AnalisisYCuracion/master/input/wtageinf.csv"
s = requests.get(url).content
dataset = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```
dataset.head()
```

```
dataset_male = dataset[dataset['Sex'] == 1][['Agemos','P3', 'P5', 'P10', 'P25', 'P50', 'P75']]
```

```
dataset_female = dataset[dataset['Sex'] == 2][['Agemos','P3', 'P5', 'P10', 'P25', 'P50', 'P75']]
```

```
dataset_male = dataset_male.melt('Agemos', var_name='P',  value_name='vals')
```

```
# Decimamos los datos para clarificar la visualización
g = sns.catplot(x="Agemos", y="vals", hue='P', data=dataset_male[0::2])
g.set_xticklabels(rotation=45)
plt.title('Niños')
```

```
dataset_female = dataset_female.melt('Agemos', var_name='P',  value_name='vals')
```

```
dataset_female.head()
```

```
g = sns.catplot(x="Agemos", y="vals", hue='P', data=dataset_female[0::2])
g.set_xticklabels(rotation=45)
plt.title('Niñas')
```
